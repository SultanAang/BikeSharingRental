### Cleaning Data

- dteday : date
- season : season (1:winter, 2:spring, 3:summer, 4:fall)
- yr (kita akan ubah menjadi year agar mudah di pahami) : year (0: 2011, 1:2012)
- mnth (kita akan ubah menjadi month agar mudah di pahami): month ( 1 dampai 12)
- hr (kita akan ubah menjadi hour agar mudah di pahami): hour (0 dampai 23)
- holiday : Mengindikasikan apakah hari itu merupakan hari libur (1 jika ya, 0 jika tidak).
- weekday : day of the week
- workingday : Menunjukkan apakah hari itu merupakan hari kerja (1 jika ya, 0 jika tidak).
- weathersit (kita akan ubah menjadi weather_situation agar mudah di pahami) :
1: Clear
2: Mist
3: Light Snow
4: Heavy Rain
- temp :  Merupakan suhu yang dinormalisasi dalam skala Celsius. Nilainya dibagi menjadi 41 (maksimal)
- atemp : Menyatakan suhu perasaan yang dinormalisasi dalam skala Celsius. Nilainya dibagi menjadi 50 (maksimal).
- hum (kita akan ubah menjadi humidity agar mudah di pahami) : Merupakan tingkat kelembaban yang dinormalisasi. Nilainya dibagi menjadi 100 (maksimal).
- windspeed (kita akan ubah menjadi wind_speed agar mudah di pahami): Menunjukkan kecepatan angin yang dinormalisasi. Nilainya dibagi menjadi 67 (maksimal).
- casual: pengendara Kasual
- registered: pengendara Terdaftar
- cnt (kita akan ubah menjadi count_cr agar mudah di pahami): Merupakan jumlah total sepeda yang disewakan, termasuk pengguna kasual dan terdaftar.

Mengganti nama kolom agar lebih mudah dibaca di dataset day_df

```python
day_df.rename(columns={'yr':'year','mnth':'month', 'weathersit':'weather_situation', 'windspeed':'wind_speed','cnt':'count_cr','hum':'humidity'},inplace=True)
```

Mengganti nama kolom agar lebih mudah dibaca di dataset hour_df

```python
hour_df.rename(columns={'yr':'year','mnth':'month','hr':'hours', 'weathersit':'weather_situation','windspeed':'wind_speed','cnt':'count_cr','hum':'humidity'},inplace=True)
```

Mengkonversi isi kolom agar mudah dipahami

```python
# mengubahtipe data dteday yang tadinya 'object' akan kita ubah menjadi 'datetime'
hour_df['dteday'] = pd.to_datetime(hour_df['dteday'])
day_df['dteday'] = pd.to_datetime(day_df['dteday'])

# mengambil hari kerja pada setiap tanggal yang sesuai
hour_df['weekday'] = hour_df['dteday'].dt.day_name()
day_df['weekday'] = day_df['dteday'].dt.day_name()

# mengambil bulan dari setiap tanggal yang sesuai
hour_df['mnth'] = hour_df['dteday'].dt.month_name()
day_df['mnth'] = day_df['dteday'].dt.month_name()

# mengambil tahun dari setiap tanggal yang sesuai
hour_df['yr'] = hour_df['dteday'].dt.year
day_df['yr'] = day_df['dteday'].dt.year

# konversi weather_situation menjadi: 1:Clear, 2:Misty, 3:Light_RainSnow 4:Heavy_RainSnow
day_df.weather_situation.replace((1,2,3,4), ('Clear','Misty','Light_rainsnow','Heavy_rainsnow'), inplace=True)
hour_df.weather_situation.replace((1,2,3,4), ('Clear','Misty','Light_rainsnow','Heavy_rainsnow'), inplace=True)

# konversi season menjadi: 1:Winter, 2:Fall, 3:Summer, 4:Spring
day_df.season.replace((1,2,3,4), ('Winter', 'Spring', 'Summer', 'Fall'), inplace=True)
hour_df.season.replace((1,2,3,4), ('Winter', 'Spring', 'Summer', 'Fall'), inplace=True)
```

```python
hour_df.head(3)
```

```python
day_df.head(3)
```

```python
# memperbaiki nilai temp, atemp, hum, dan kecepatan angin ke nilai sebelum dinormalisasi di hour_df
hour_df['temp'] = hour_df['temp']*41
hour_df['atemp'] = hour_df['atemp']*50
hour_df['humidity'] = hour_df['humidity']*100
hour_df['wind_speed'] = hour_df['wind_speed']*67
```

```python
hour_df.head(8)
```

```python
# Perbaiki nilai temp, atemp, hum, dan kecepatan angin ke nilai sebelum dinormalisasi di day_df
day_df['temp'] = day_df['temp']*41
day_df['atemp'] = day_df['atemp']*50
day_df['humidity'] = day_df['humidity']*100
day_df['wind_speed'] = day_df['wind_speed']*67
```

### Boxplots of bikeshare users by season

```python
plt.figure(figsize=(16,6))

# membuat boxplot
sns.boxplot(
    x="season",
    y="count_cr",
    data=hour_df,
    palette=["lightblue", "cyan"]
)

# menambahkan label x dan y
plt.xlabel("Season")
plt.ylabel("Total Rides")
plt.title("Count of bikeshare rides by Season")


plt.show()
```

Berdasarkan gambar di atas, Anda dapat melihat bahwa penyewaan sepada paling banyak berdasarkan musim yaitu di musim panas (summer)

```python
plt.figure(figsize=(16,6))

# membuat boxplot
sns.boxplot(
    x="hours",
    y="count_cr",
    data=hour_df,
    palette=["lightblue", "cyan"]
)

# menambahkan label x dan y juga title nya
plt.xlabel("Hour")
plt.ylabel("Total Rides")
plt.title("Count of bikehsare rides by Hour")

# Show the plot
plt.show()
```

- jumlah pengguna sepeda yang rendah ada dari tengah malam hingga fajar, yaitu pukul 00.00 hingga pukul 07.00
- Jumlahnya mulai meningkat secara drastis di pagi hari, dan puncaknya pada pukul 08.00
- Jumlahnya mulai menurun lagi menuju siang hari, tetapi pukul 16.00 jumlah pengguna sepeda mulai meningkat lagi dan mencapai puncaknya lagi pada pukul 17.00
- Setelah itu jumlah pengguna seped menurun kembali hingga tengah malam

```python
plt.figure(figsize=(16,6))

# membuat boxplot
sns.boxplot(
    x="weather_situation",
    y="count_cr",
    data=hour_df,
    palette=["lightblue", "cyan"]
)

# menambahkan label x dan y
plt.xlabel("Hour")
plt.ylabel("Total Rides")
plt.title("Count of bikehsare rides by Weather")


plt.show()
```

Jumlah pengguna sepeda yang paling sedikit ada di musim dingin, dan paling yang diminati yaitu di musim panas.

```python
hourly_count= hour_df.groupby(['hours', 'workingday']).agg({
    "count_cr": "sum"
}).reset_index()

sns.lineplot(data=hourly_count,x="hours",y="count_cr",hue="workingday",palette="viridis",marker="o")

plt.title("Total sepeda berdasarkan jam di hari Kerja")
plt.xlabel("Bulan")
plt.ylabel(None)
plt.legend(title="workingday", loc="upper right")
plt.tight_layout()
plt.show()
```

- Berdasarkan gambar diatas, orang yang menyewa dihari kerja (1) pada dini jumlah nya menurun, lalu mulai naik drastis pada pukul 05.00 sampai sekitar pukul 08.00 lalu menurun drastis di pukul 10.00, lalu naik kembali dipukul 15.00 dan menurun kembali hingga dini hair
- untuk orang yang menyewa yang bukan hari kerja (0) pada dini hari mulai menurun hingga pukul 05.00 lalu menaik walaupun menaiknya tidak signifikan seperti di hari kerja

```python
plt.figure(figsize=(8,6))
sns.barplot(x='workingday', y='count_cr', data=hour_df, palette="Set2")
plt.title('Pola Penggunaan Sepeda pada Hari Kerja vs Akhir Pekan')
plt.xlabel('Hari Kerja (1) vs Akhir Pekan (0)')
plt.ylabel('Rata-rata Pengguna Sepeda')
plt.xticks([0, 1], ['Akhir Pekan', 'Hari Kerja'])
plt.show()
```

Rata-rata jumlah penyewa sepeda yang paling tinggi itu ada di hari kerja, dan perbedaanna pun tidak jauh berbeda dengan akhir pekan





```python
# Visualisasi rata-rata penggunaan sepeda berdasarkan hari dalam minggu
plt.figure(figsize=(12,6))
sns.lineplot(x='weekday', y='count_cr', data=hour_df, color='green')
plt.title('Pola Penggunaan Sepeda Berdasarkan Hari dalam Minggu')
plt.xlabel('Hari dalam Minggu (0=Senin, 6=Minggu)')
plt.ylabel('Jumlah Pengguna Sepeda')
plt.grid(True)
plt.show()
```

Dari gambar di atas, titik paling terendah yaitu di hari minggu (sunday), dan mulai menaik di hari biasa (senin). Titik paling tertinggi nya yaitu di hari jumat(Friday) dan ketika menuju weekend, penyewa sepeda mulai menurun kembali

```python
plt.figure(figsize=(10,6))
sns.lineplot(x='season', y='registered', data=hour_df, color='blue', label='Terdaftar')
sns.lineplot(x='season', y='casual', data=hour_df, color='orange', label='Kasual')
plt.title('Pola Penggunaan Sepeda Kasual vs Terdaftar Berdasarkan Musim')
plt.xlabel('Musim (1=Musim Dingin, 2=Musim Semi, 3=Musim Panas, 4=Musim Gugur)')
plt.ylabel('Rata-rata Pengguna Sepeda')
plt.legend()
plt.show()
```

- Kesamaan antara penyewa Kasual dengan penyewa terdaftar yaitu dimusim.
di musim dingin (Winter) para penyewa sepeda berada di titik terendah dan mulai naik di musim semi (Spring) dan puncak nya berada di musim panas (summer) dan mulai turun kembali di musim gugur (Fall)
- Untuk perbedaannya hanya ada di jumlah rata-ratanya. Penyewa terdaftar lebih banyak dibanding penyewa kasual

```python
plt.figure(figsize=(10,6))
sns.barplot(x='weather_situation', y='registered', data=hour_df, color='orange', label='Terdaftar')
sns.barplot(x='weather_situation', y='casual', data=hour_df, color='blue', label='Kasual')
plt.title('Pola Penggunaan Sepeda Kasual vs Terdaftar Berdasarkan Kondisi Cuaca')
plt.xlabel('Kondisi Cuaca (1=Cuaca Cerah, 2=Berawan, 3=Hujan, 4=Hujan Lebat)')
plt.ylabel('Rata-rata Pengguna Sepeda')
plt.legend()
plt.show()
```

Dari Bagan di atas terlihat bahwa bagian cuaca cerah(Clear) lebih tinggi bagian yang lainnya

### Pertanyaan 3 : di bulan dan tahun berapakah penyewaan sepeda itu sedang tinggi dan di bulan apakah penyewaan sepeda itu sedang turun

```python

monthly_counts = hour_df.groupby(by=["month","year"]).agg({
    "count_cr": "sum"
}).reset_index()

sns.lineplot(data=monthly_counts,x="month",y="count_cr",hue="year",palette="viridis",marker="o")

plt.title("Total sepeda berdasarkan Bulan dan tahun")
plt.xlabel("Bulan")
plt.ylabel(None)
plt.legend(title="Tahun", loc="upper right")
plt.tight_layout()
plt.show()
```

Dari gambar diatas merupakan total penyewa sepeda per bulan berdasarkan tahun</br>
- Di Tahun 2011(0) total penyewa tertinggi berada dibulan Juni diangka sekitar 125000-150000 dan menurun ke angka 75000-100000
- Di Tahun 2012(1) total penyewa tertinggi berada dibulan September diangka sekitar 200000-225000 dan menurun ke angka 125000-150000
